```python
#code to clean the texts up a bit 
limit = 0

def cleaner(df):
    #Regex pattern for only alphanumeric, hyphenated text with 3 or more chars
    pattern = re.compile(r"[A-Za-z0-9-.\-]{1,30}")
    df['cleaned_content'] = df['cleaned_content'].str.findall(pattern).str.join(' ')
    if limit > 0:
        return df.iloc[:limit, :].copy()
    else:
        return df

cleaner(dt)
```

```python
dt = dt.replace({np.nan: None})

corpus_art = dt[['cleaned_content','id']]
corpus_art.drop_duplicates(subset = ["cleaned_content"], keep = 'first', inplace = True)

corpus_art = corpus_art.rename(columns={"cleaned_content": "content_", "index": "index_"})

corpus_art_list = list(corpus_art['content_'])


tfidf = TfidfVectorizer(token_pattern = r"(?u)\b\w+\b", max_features= 1000)
x_art = tfidf.fit_transform(corpus_art_list)

df_tfidf_art = pd.DataFrame(x_art.toarray(), columns=tfidf.get_feature_names())

corpus_art = pd.merge(corpus_art, df_tfidf_art, left_index=True, right_index=True, how='left')
df_tfidf_art_updated = pd.merge(dt['cleaned_content'], corpus_art, left_on='cleaned_content',
                                right_on='content_', how='left')

df_tfidf_art_updated.head(3)
```
